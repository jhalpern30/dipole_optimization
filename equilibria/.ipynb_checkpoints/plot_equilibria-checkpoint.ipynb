## Coils initialization

```python
# Generate some TF coils.
base_tf_curves_cylindrical = create_equally_spaced_cylindrical_curves(ntf, surf.nfp, stellsym=True, R0=HBT_R0, minor_r= TF_R1,numquadpoints=32)
tf_curves_cylindrical = apply_symmetries_to_curves( 
    base_curves=base_tf_curves_cylindrical,
    nfp=surf.nfp,
    stellsym=surf.stellsym
)

# We define the currents with 1A, so that our dof is order unity. We
# then scale the current by a scale_factor of 1E6.
scale_factor = 2.5E5
base_tf_currents = [Current(1) for c in base_tf_curves_cylindrical]
tf_currents = apply_symmetries_to_currents( 
    base_currents=base_tf_currents, 
    nfp=surf.nfp,
    stellsym=surf.stellsym
)

base_tf_coils = [Coil(curve, ScaledCurrent(current, scale_factor)) for curve, current in zip(base_tf_curves_cylindrical, base_tf_currents)]
tf_coils = [Coil(curve, ScaledCurrent(current, scale_factor)) for curve, current in zip(tf_curves_cylindrical, tf_currents)]
```

```python
# number of WP coils = ntoroidal * npoloidal * nfp * 2 (stellsym)
nwptot = 2 * npoloidal * ntoroidal * surf.nfp
base_wp_curves = create_equally_spaced_windowpane_grid(
    ntor=ntoroidal,
    npol=npoloidal,
    nfp=surf.nfp,
    stellsym=surf.stellsym,
    R0=HBT_R0,
    a=HBT_a,
    R1=dipole_radius,
    order=1,
    numquadpoints=64
)

wp_curves = apply_symmetries_to_curves( 
    base_curves=base_wp_curves,
    nfp=surf.nfp,
    stellsym=surf.stellsym
)

base_wp_current = [Current(1) for c in base_wp_curves]
wp_currents = apply_symmetries_to_currents( 
    base_currents=base_wp_current, 
    nfp=surf.nfp,
    stellsym=surf.stellsym
)

wp_scale_factor = 1E4 # We again apply the trick to have a dof of order unity
wp_coils = [Coil(curve, ScaledCurrent(current, wp_scale_factor)) for curve, current in zip(wp_curves, wp_currents)]
base_wp_coils = [Coil(curve, ScaledCurrent(current, wp_scale_factor)) for curve, current in zip(base_wp_curves, base_wp_current)]

for ii in range(len(base_wp_curves)):
    base_wp_coils[ii].name = f'WP_{ii+1}'

for ii in range(ntf):
    base_tf_coils[ii].name = f'TF_coil_{ii+1}'
    
# the base TF coils are post TF optimization
base_coils = base_tf_coils + base_wp_coils
coils = tf_coils + wp_coils
```

## Coil Optimization 

```python
# set tf coil dofs
for c in base_tf_curves_cylindrical:
    c.fix_all()

for c in base_tf_curves_cylindrical:
    c.unfix('R0')
for c in base_tf_curves_cylindrical:
   c.unfix('r_rotation')
#for c in base_tf_curves_cylindrical:
#   c.unfix('z_rotation')
```

```python
def classify_current(item, threshold):
    # the first ntf currents correspond to TF coils, rest are dipoles
    # Extract the number after 'Current' and before ':'
    if item.startswith('Current'):
        current_num = int(item.split(':')[0].replace('Current', ''))
        # Return 0 if the number is less than threshold, otherwise return 1
        return 0 if current_num < threshold else 1    
    else:
        return 0
```

```python
def derivativeJcp(current, CURRENT_THRESHOLD):
    #manually calculate derivative of current penalty
    diff = np.abs(current) - CURRENT_THRESHOLD
    mask = diff > 0
    grad = np.where(current > 0, 2 * (current - CURRENT_THRESHOLD), 2 * (current + CURRENT_THRESHOLD))
    return grad * mask
```

```python
# Define objective function
bs = BiotSavart( coils )
bs2 = BiotSavart(tf_coils)

# Number of iterations to perform:
MAXITER = 10000

# Threshold and weight for the coil-to-coil distance penalty in the objective function:
# using peak field approximation Bmax = mu0 * I / 2 / R
CC_THRESHOLD = 0.1
CC_WEIGHT = 100000
Jccdist = CurveCurveDistance(tf_curves_cylindrical, CC_THRESHOLD, num_basecurves=ntf)

# Threshold and weight for the coil-to-surface distance penalty in the objective function:
CS_THRESHOLD = 0.05
CS_WEIGHT = 100000
#Jcsdist = CurveSurfaceDistance(base_tf_curves_cylindrical, surf, CS_THRESHOLD)
Jcsdist = CurveSurfaceDistance(base_tf_curves_cylindrical, VV, CS_THRESHOLD) # what about vv to coil distance not plasma surface?

#Adding in toroidal flux in order to keep the current coils optimizable ???
TF_WEIGHT = 100
TF_THRESHOLD = 0.01
Tf = ToroidalFlux(surf,bs2)

# Penalize any links between TF curves (interlocking coils not allowed)
LINKNUM_WEIGHT = 1
LINKNUM_THRESHOLD = 0.1
linkNum = LinkingNumber(tf_curves_cylindrical)

Jf = SquaredFlux(surf, bs, definition='local')

# Current penality threshold and weight
CURRENT_THRESHOLD = dip_to_TF_peak_B_ratio * scale_factor * dipole_radius / TF_R1
CURRENT_WEIGHT = 1E-5
#use this to only set dJ for dipole current dofs
dJscale = [classify_current(dof_name, ntf+1) for dof_name in Jf.dof_names]

tf_opt_params = ['CoilCoil', CC_THRESHOLD, CC_WEIGHT, 'CoilVV', CS_THRESHOLD, CS_WEIGHT, 
                 'Toroidal Flux', TF_WEIGHT, TF_THRESHOLD, 'Link Num', LINKNUM_WEIGHT, LINKNUM_THRESHOLD,
                 'Current Threshold', CURRENT_WEIGHT, CURRENT_THRESHOLD]

# Set total objective
curve_distance = CC_WEIGHT * Jccdist 
surface_distance = CS_WEIGHT* Jcsdist
JF = Jf \
    + CC_WEIGHT * Jccdist \
    + CS_WEIGHT* Jcsdist  + TF_WEIGHT *  QuadraticPenalty(Tf, TF_THRESHOLD, "min") #+ LINKNUM_WEIGHT * QuadraticPenalty(linkNum, LINKNUM_THRESHOLD,'max')

print(f'Current Threshold= {CURRENT_THRESHOLD:.2e} A'), 
print('Names of the dofs:', JF.dof_names)

def fun(dofs):

    # this is a hardcoded current threshold penalty
    Jcp = np.sum(dJscale * np.array([np.maximum(np.abs(dofs[i]*wp_scale_factor) - CURRENT_THRESHOLD, 0)**2 for i in range(len(dofs))]))
    dJcp = dJscale * np.array([derivativeJcp(dofs[i]*wp_scale_factor, CURRENT_THRESHOLD) for i in range(len(dofs))])
    
    JF.x = dofs
    J = JF.J() + CURRENT_WEIGHT * Jcp
    grad = JF.dJ() + CURRENT_WEIGHT * dJcp
    jf = Jf.J()
    tf = Tf.J()
    
    BdotN = np.mean(np.abs(np.sum(bs.B().reshape((nphi, ntheta, 3)) * surf.unitnormal(), axis=2)))
    outstr = f"J={J:.1e}, Jf={jf:.1e}, ⟨B·n⟩={BdotN:.1e}, Jcp ={Jcp:.1e}"
    outstr += f", C-C-Sep={Jccdist.shortest_distance():.2f}, C-S-Sep={Jcsdist.shortest_distance():.2f}"
    outstr += f", Toroidal Flux={tf:.1e} ║∇J║={np.linalg.norm(grad):.1e}"
    print(outstr)
    return J, grad
```

```python
theta = surf.quadpoints_theta
phi = surf.quadpoints_phi
ntheta = theta.size
nphi = phi.size

Bdotn = np.sum(bs.B().reshape((nphi, ntheta, 3)) * surf.unitnormal(), axis=2)
modB = bs.AbsB().reshape((nphi,ntheta))
BdotN_norm = Bdotn / modB
BdotN2 = BdotN_norm**2
BdotN2_avg = simpson(simpson(BdotN2 * np.linalg.norm(surf.normal(), axis = 2), theta), phi) # this is the BdotN^2 times the scaled current in each coil, summed over the coils, and averaged over phi and theta

print('Initial Integrated BdotN^2 / |B|^2: ' + str(BdotN2_avg))
initial_avgBdotN2 = BdotN2_avg
initial_maxBdotN2 = np.max(BdotN2)


fig, ax = plt.subplots(figsize=(12,5))
c = ax.contourf(phi,theta,np.transpose(Bdotn / modB))
plt.colorbar(c)
ax.set_title(r'$Initial ~ \mathbf{B}\cdot\hat{n} / |B|$ ')
ax.set_ylabel(r'$\theta$')
ax.set_xlabel(r'$\phi$')
plt.savefig(dir + '/InitialBdotN')
```
